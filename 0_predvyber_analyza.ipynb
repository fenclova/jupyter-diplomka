```python
# PODMÍNKA: povolit problem pri hledani cesty? 0 = nepovoleno
find_path_problem_podminka = 0

# PODMÍNKA: podíl ploch po rozdělení čtverce vodním tokem
nejlepsi_podil_ploch_max = 5 

# minimální počet průsečíků vodního toku a mřížky
pocet_pruseciku_min = 10

# PODMÍNKA: délka železnice 
zeleznice_delka_min = 2*500 # v mape generalizace 1 : 50 000 linie 2 cm
zeleznice_delka_max = 15000

# PODMÍNKA: minimální délka rozvodncie III. řádu
rozvodniceIII_delka_min = 2*250 # v mapě linie 2 cm

# PODMÍNKA: výskyt vodní plochy, minimální délka
vodni_plohy_min = 1*62500 # 1cm2 v mapě 1 : 25000
vodni_plohy_max = 16*62500 # 16cm2 v mapě = malý čtvereček !!! zde pozor > řeka = vodní plocha

# PODMÍNKA: maximalni povolena rozloha skal a povrchové těžby
relief_rozloha_max = 2*62500 # 2cm2 v mapě

# PODMÍNKA: výskyt zástavby
zastavba_rozloha_min = 1*62500 # 1cm2 v mapě

vyber1 = df[(df.nejlepsi_podil_ploch >=1) & (df.nejlepsi_podil_ploch <= nejlepsi_podil_ploch_max) &
   (df.pocet_pruseciku >= pocet_pruseciku_min) &
    (df.zeleznice_delka >= zeleznice_delka_min) & (df.zeleznice_delka <= zeleznice_delka_max) &
   (df.rozvodniceIII_delka >= rozvodniceIII_delka_min) &
   (df.relief_rozloha <= relief_rozloha_max) &
   (df.vodni_plohy_rozloha > vodni_plohy_min) & #(df.vodni_plohy_rozloha <= vodni_plohy_max) &
   (df.zastavba_rozloha > zastavba_rozloha_min)].set_index('ID')

# bez podminky o problemu pri hledani vodniho toku + 23
# bez podminky vodni plochy + 81
```
